# Predictive Modeling
## Data Prepration
- Divide the data into features (X) and targets (y).
- Encode categorical variables.
- Divided into training and testing units.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#defining features and target var
X = df.drop(columns=['customerID', 'Churn'])
#varify
print('X:',X)
print('df:',df.head())
```

X: This line creates an attribute matrix X by removing the customerID and Churn columns from the data set. The customerID is discarded because it is only an identifier and does not provide any useful information for predicting churn. The Churn column is dropped because it is a target variable.

```python
y = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
```

y: This line creates the target vector y by converting the Churn column to binary values (1 for "Yes" and 0 for "No"). This is done using a lambda function that applies the variable to each value in the Churn column.

```python
#varify
print('y:',y)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

**pd.get_dummies:** This function converts categorical variables in the feature matrix X into numerical values by creating dummy variables (one-hot encoding which is used when categorical variables do not have an ordinal relationship, ensuring that the model does not assume any unintended hierarchy among the categories). The drop_first=True argument drops the first category of each categorical variable to avoid multicollinearity (a situation where features are highly correlated, which can affect the model’s performance).

```python
# Splitting the Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

- **test_size=0.3:** 30% of the data is reserved for testing, the remaining 70% for training.
- **random_state=42**: This ensures that the split is reproducible (the same split will occur every time the code is run).
- **stratify=y:** ensure that the class distribution of y (target variable) is preserved in both training and testing sets.

```python
#Standardizing Features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

- **StandardScaler:** Imagine you have a bunch of different features (like age, income, or height) that are on different scales—some might be in the thousands, while others are just in the tens. The StandardScaler helps even out these differences by adjusting the data so that each feature has a similar scale. This is particularly important for certain models, like Logistic Regression, which can be influenced by the scale of the data.

- **scaler.fit_transform(X_train):** This step is like learning the patterns in the training data. The scaler first figures out the average value (mean) and how spread out the data is (standard deviation) for each feature in the training set. Then, it uses this information to adjust (or "scale") the training data so that everything is on a similar scale.

- **scaler.transform(X_test):** Now that the scaler knows how to adjust the data, it applies this same adjustment to the test data. It uses the mean and standard deviation calculated from the training set to ensure that the test data is scaled in exactly the same way. This consistency is key for making accurate predictions when you test your model.

## Model Training

```python
print(X.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

# Creating an imputer object with a strategy to fill missing values with the mean (or other strategies like 'median', 'most_frequent')
imputer = SimpleImputer(strategy='mean')

# Fiting the imputer on the training data and transform both the training and testing data
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Training Logistic Regression model
#
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

```

- **random_state=42:** Ensures that the randomness in the algorithm (if any) is reproducible

```python
# Making predictions on the test set
y_pred = model.predict(X_test)

```

- **model.predict(X_test):** After training the model on your data, you now want to see how well it performs on new, unseen data (the test set). The model takes in the features from the test set and predicts whether each customer is likely to churn (leave) or not. The predictions are stored in y_pred, with each entry being a 0 (indicating the customer is not predicted to churn) or a 1 (indicating they are predicted to churn).

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

- **accuracy_score:** Calculates the accuracy of the model, which is the proportion of correctly predicted instances out of all instances.
- **precision_score:** Calculates the precision of the model, which is the proportion of true positive predictions out of all positive predictions made by the model. Precision is important in situations where the cost of a false positive is high.
- **recall_score:** Calculates the recall (or sensitivity) of the model, which is the proportion of true positives out of all actual positives. Recall is important when the cost of a false negative is high.
- **f1_score:** Calculates the F1 score, which is the harmonic mean of precision and recall. It provides a balance between precision and recall, especially when you have an imbalanced dataset.

```python
#Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", conf_matrix)
```

- **confusion_matrix:** Generates a confusion matrix, which is a table showing the true positives, false positives, true negatives, and false negatives. This gives a detailed view of how the model's predictions compare to the actual labels.

```python
#Classification Report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", class_report)
```
